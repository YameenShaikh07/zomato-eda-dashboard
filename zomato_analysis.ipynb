```python
import pandas as pd

# Try reading the CSV file using 'latin-1' encoding
df = pd.read_csv('/content/zomato.csv', encoding='latin-1')
# If 'latin-1' doesn't work, try other encodings like 'ISO-8859-1', 'cp1252', etc.
# df = pd.read_csv('/content/zomato.csv', encoding='ISO-8859-1')

df.head()
df.info()
df.describe()
```

```python
df.drop([
    'Restaurant ID', 'Address', 'Locality', 'Locality Verbose',
    'Longitude', 'Latitude', 'Switch to order menu'
], axis=1, inplace=True)
```

```python
df = df[df['Cuisines'].notnull()]
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

```

```python
df['cuisines'] = df['cuisines'].apply(lambda x: str(x).split(',')[0].strip())
df['has_table_booking'] = df['has_table_booking'].map({'Yes': 1, 'No': 0})
df['has_online_delivery'] = df['has_online_delivery'].map({'Yes': 1, 'No': 0})
```

```python
df['price_category'] = pd.cut(df['average_cost_for_two'],
                              bins=[0, 300, 800, 2000, 5000],
                              labels=['Budget', 'Mid-range', 'Expensive', 'Luxury'])
```

```python
# Save the cleaned dataset to a CSV file
df.to_csv('/content/cleaned_zomato.csv', index=False)
```
